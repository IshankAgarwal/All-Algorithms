```python
import random
import time as t
%matplotlib inline
import matplotlib.pyplot as plt
q=[]
r=[]
z=0
avtimelist_b=[]
avtimelist_i=[]


timelist_b=[]
timelist_i=[]

def bubblesort(list1):
	#print(list1)

	for i in range(len(list1)):
		for j in range(len(list1)-1):
			if(list1[j]>list1[j+1]):
				list1[j],list1[j+1]=list1[j+1],list1[j]

	return (list1)


#list2=[10,25,8,11,37,6,19,28]

def insertionsort(list2):
	
	for i in range(1, len(list2)): 
  
	        key = list2[i] 
	        j = i-1
	        while j >=0 and key < list2[j] : 
	                list2[j+1] = list2[j] 
	                j -= 1
	        list2[j+1] = key
	 
	return (list2)

#print (insertionsort(list2))


x=[1000,5000,10000,15000,20000,25000,30000]
#print(x)
for r in range(len(x)):

	for n in range(0,3):

		for k in range(x[z]):
			p=random.randint(0,100)
			q.append(p)
		r=q[:]

		a=t.time()
		bubblesort(q)
		b=t.time()
		
		c=t.time()
		insertionsort(r)
		d=t.time()
		
		timediff_b=b-a

		timediff_i=d-c
		
		timelist_b.append(timediff_b)
		timelist_i.append(timediff_i)
		del q[:]
		del r[:]

		#print(timelist,end="")

	avtimelist_b.append((timelist_b[0]+timelist_b[1]+timelist_b[2])/3)

	avtimelist_i.append((timelist_i[0]+timelist_i[1]+timelist_i[2])/3)


	del timelist_b[:]
	del timelist_i[:]
	
	timelist_b=[]
	timelist_i=[]

	z=z+1


plt.plot(x,avtimelist_b,'r')
plt.plot(x,avtimelist_i,'b')
plt.show()

# RED CURVE  = BUBBLE SORT
# BLUE CURVE = INSERTION SORT
```
